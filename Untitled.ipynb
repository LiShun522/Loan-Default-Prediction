```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("data/train.csv")
df = pd.DataFrame(data)
```

```python
df.head(20)
```

```python
# 設定異常值閾值（可以使用 99% 分位數或明顯異常的數據，如 99999999）
threshold = df["Current Loan Amount"].quantile(0.99)
df["is_outlier"] = df["Current Loan Amount"] > threshold

# 計算兩組的違約率
default_rates = df.groupby("is_outlier")["Credit Default"].mean()
print(default_rates)
```

```python
# upper_limit = df["Current Loan Amount"].quantile(0.99)

# # 過濾掉異常值
# df_filtered = df[df["Current Loan Amount"] < upper_limit]

# 重新繪製直方圖
sns.displot(df, x="is_outlier", bins=10, hue="Credit Default")
# 旋轉 X 軸標籤，確保不重疊
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# 移除異常值（設定閾值，如 99% 分位數）
threshold = df["Current Loan Amount"].quantile(0.99)
df_cleaned = df[df["Current Loan Amount"] < threshold]

# 計算相關係數矩陣
print(df_cleaned[["Current Loan Amount", "Credit Score", "Credit Default"]].corr())
print(df[["Current Loan Amount", "Credit Score", "Credit Default"]].corr())
```

```python
# 先排除異常值，計算每個類別的中位數
grouped_median = df[df["Current Loan Amount"] != 99999999.0].groupby(["Purpose", "Term"])["Current Loan Amount"].median()

# 檢視不同類別的中位數
print(grouped_median)
```

```python
# 先確認異常值
mask = df["Current Loan Amount"] == 99999999.0

# 根據 (Purpose, Term) 來填補異常值
df.loc[mask, "Current Loan Amount"] = df.loc[mask, ["Purpose", "Term"]].apply(
    lambda x: loan_median_dict.get((x["Purpose"], x["Term"]), df["Current Loan Amount"].median()), axis=1
)
```

```python
print("填補後仍然有異常值嗎？", (df["Current Loan Amount"] == 99999999.0).sum())
```
